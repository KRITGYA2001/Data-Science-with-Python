```python
#using shorthand form 
#linear graph
x=np.array([10,20,30,40,50])
y=np.array([10,20,30,40,50])
plt.plot(x,y,"o:g")
plt.show()
```

```python
#using shorthand form 
#curve graph
x=np.array([10,20,30,40,50])
y=np.array([10,2,30,4,50])
plt.plot(x,y,"*:r")
plt.show()
```

```python
#in shorthand form "o:r" := shows line specification 
```

```python
#Line specification
#linear graph
x=np.array([1,2,3,4,5,6])
y=np.array([10,20,30,40,50,60]) 
plt.plot(x,y,"o--r")
plt.show()
```

```python
#Line specification
#curve graph
x=np.array([1,2,3,4,5,6])
y=np.array([10,2,30,4,50,6]) 
plt.plot(x,y,"o--y")
plt.show()
```

```python
#ms = marker size
#mec = marker edge color
#mfc = marker face color
```

```python
#We can change marker size by using ms="" function
#mec function used to color the edges
#linear graph
x=np.array([1,2,3,4,5,6])
y=np.array([10,20,30,40,50,60])         
plt.plot(x,y,marker="*",ms=10,mec='y')
plt.show()
```

```python
#curve graph
x=np.array([1,2,3,4,5,6])
y=np.array([10,2,30,4,50,6])         
plt.plot(x,y,marker="x",ms=10,mec='y')
plt.show()
```

```python
#we can make marker face color by using mfc=""
#linear graph
x=np.array([1,2,3,4,5,6])
y=np.array([10,20,30,40,50,60])
plt.plot(x,y,marker="o",ms=10,mec="black",mfc='r')
plt.show()
```

```python
#curve graph
x=np.array([1,2,3,4,5,6])
y=np.array([10,2,30,4,50,6])
plt.plot(x,y,marker="o",ms=10,mec="cyan",mfc='pink')
plt.show()
```

```python
#we can use color code instead of color name
#mfc = #456ABC (some color)   #4CAF56
#linear graph
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,marker="o",ms=10,mfc='#4CAF56')
plt.show()
```

```python
#curve graph
x=np.array([1,2,3,4,5])
y=np.array([10,2,30,4,50])
plt.plot(x,y,marker="o",ms=10,mfc='#3ACF10')
plt.show()
```

```python
#we can change the graph line by using linestyle function
#dotted #dashed #linestyle="--"
#linear graph
x=np.array([1,2,3,4,5])
y=np.array([20,40,60,80,100])
plt.plot(x,y,linestyle="dotted")   
plt.show()  
```

```python
#giving labels of x and y axis by using xlabel and ylabel
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,linestyle="dashed",color="r",linewidth="5") 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show() 
```

```python
#creating own font family style
#we can use our created font by fontdict in xlabel and ylabel
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
a={"family":"DejaVu Sans"}
b={"family":"times new roman"}
plt.plot(x,y,linestyle="dashed",color="r",linewidth="5") 
plt.xlabel("X-axis",fontdict=b)
plt.ylabel("Y-axis",fontdict=a)
plt.show() 
```

```python
#assiging the title of graph
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,linestyle="dashed",linewidth="1",marker="*",color="black",ms=10)
plt.title("This is the sample title of the graph",loc="left")
plt.show()
```

```python
#grid system
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y,linestyle="dashed",linewidth="1",marker="*",color="black",ms=10)
plt.title("This is the sample title of the graph",loc="left")
plt.grid()
plt.show()
```

# Bar Graph

```python
x=np.array([10,60,6,45,50,90,1])
y=np.array([1,2,3,4,5,6,7])
z=np.array(["green","brown",'black',"red","yellow","orange","blue"])
plt.bar(x,y)
plt.show()
```

```python
x=np.array([10,60,6,45,50,90,1])
y=np.array([1,2,3,4,5,6,7])
z=np.array(["green","brown",'black',"red","yellow","orange","blue"])
plt.bar(y,x)
plt.show()
```

```python
x=np.array(["A","B",'C',"D","E","F","G"])
y=np.array([1,2,3,4,5,6,7])
plt.bar(x,y)
plt.show()
```

```python
#bars on horizontal line
x=np.array(["A","B",'C',"D","E","F","G"])
y=np.array([1,2,3,4,5,6,7])
plt.barh(x,y)
plt.show()
```

```python
#changing the color of bars
```

```python
x=np.array(["A","B",'C',"D","E","F","G"])   #independent quantity
y=np.array([1,2,3,4,5,6,7])           #dependent quantity
plt.bar(x,y,color="red")
plt.show()
```

```python
#width of bar
x=np.array(["A","B",'C',"D","E","F","G"])
y=np.array([1,2,3,4,5,6,7])
plt.bar(x,y,color="red",width=0.3)
plt.show()
```

```python
x=np.array(["A","B",'C',"D","E","F","G"])
y=np.array([1,2,3,4,5,6,7])
plt.barh(x,y,height=0.3,color="g")
plt.show()
```

# Pie charts

```python
x=np.array([25,25,15,35])
plt.pie(x)
plt.show()
```

```python
x=np.array([25,25,15,35])
y=["apple","mango","gauva","pier"]
plt.pie(x,labels=y)
plt.show()
```

```python
#changing the start angle
x=np.array([25,25,15,35])
y=["apple","mango","gauva","pier"]
plt.pie(x,labels=y,startangle=90)
plt.show()
```

```python
#separate the sectors in pie chart
x=np.array([25,25,15,35])
y=["apple","mango","gauva","pier"]
z=[0.2,0.3,0.1,0.1]
plt.pie(x,labels=y,startangle=90,explode=z)
plt.show()
```

```python
#3D view
x=np.array([25,25,15,35])
y=["apple","mango","gauva","pier"]
z=[0.2,0.3,0.1,0.1]
plt.pie(x,labels=y,startangle=90,explode=z,shadow=True)
plt.show()
```

```python
#to display small box of information
x=np.array([25,25,15,35])
y=["apple","mango","gauva","pier"]
z=[0.2,0.3,0.1,0.1]
plt.pie(x,labels=y,startangle=90,explode=z,shadow=True)
plt.legend()
plt.show()
```

```python
#percentage 
x=np.array([25,25,20,30])
y=["apple","mango","gauva","pier"]
z=[0.2,0.3,0.1,0.1]
plt.pie(x,labels=y,startangle=90,explode=z,shadow=True)
plt.legend()
plt.show()
```
