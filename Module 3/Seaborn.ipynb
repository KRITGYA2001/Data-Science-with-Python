Displot is used to plot a particular column 
It represents both histogram and line graph
Displot only contain one column(one value)

```python
#Uni variant (Single Variable)
#for Univariant we use distplot function
#kde = Kernal Density
file=sb.load_dataset('iris')
sb.distplot(file["sepal_length"])
plt.show()
#it print both line graph and histogram graph
```

```python
sb.distplot(file["petal_width"],kde=False)
plt.show()
#by using kde=False we can't print line only histogram function
```

```python
sb.set_style("whitegrid")
sb.distplot(file["petal_width"],hist=False)
plt.show()
#only line we can use hist=False
```

```python
jointplot is used for multiple columns to represent
it represent both histogram and point graph(by default)
```

```python
#Bi variant and multi variant
#In this we will use jointplot function
#kind = define the type of graph
sb.jointplot(x="petal_length",y="petal_width",data=file)
plt.show()
#Its create the histogram and scatter graph by default
```

```python
#We can convert scatter points into hexagonal shape
sb.jointplot(x="sepal_length",y="sepal_width",data=file,kind="hex")
plt.show()
```

```python
#Kde = density graph (line graph)
sb.jointplot(x="sepal_length",y="petal_length",data=file,kind='kde')
plt.show()
```

```python
#In pairplot 
#diag_kind = graph type
#kind = type of graph
#palette = color pattern
sb.pairplot(file,hue="species",kind="scatter",palette="brg")
plt.show()
```

```python
sb.pairplot(file,hue="species",diag_kind="hex",kind="scatter",palette="brg_r")
plt.show()
```

```python
#Instead of Scatter plot we use stripplot
sb.stripplot(x="species",y='petal_length',data=file)
plt.show()
```

```python
#There is jitter=True function to clear some point for clear view
sb.stripplot(x="species",y="petal_width",data=file,jitter=True)
plt.show()
```

```python
#jitter = False to see a point line of data
sb.stripplot(x="species",y="petal_width",data=file,jitter=False)
plt.show()
```

```python
#And too see more clearly we use swarmplot function
sb.swarmplot(x="species",y="sepal_length",data=file)
plt.show()
```

```python
#To see block diagram of points
sb.boxplot(x="species",y="sepal_width",data=file)
plt.show()
```
